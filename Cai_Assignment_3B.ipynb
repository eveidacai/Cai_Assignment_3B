```python
## Code goes here
import pandas as pd
import csv 
import re
```

```python
#merging

recweekdata_df = pd.read_csv("recweekdata.csv")
studentprofiles_df = pd.read_csv("studentprofiles.csv")
main_df = pd.merge(studentprofiles_df, recweekdata_df, how="left", left_on=["lastname"], right_on=["lastname"])
main_df = main_df.drop(columns=['firstname_y'])
main_df.rename(columns={'firstname_x':'firstname'}, inplace=True)
```

```python
#birthmonth

main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(January|Jan|1)",r"January", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(February|Feb|2)",r"February", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(March|Mar|3)",r"March", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(April|Apr|4)",r"April", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(May|May|5)",r"May", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(June|Jun|6)",r"June", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(July|Jul|7)",r"July", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(August|Aug|8)",r"August", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(September|Sep|9)",r"September", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(January0|October|Oct|10)",r"October", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(JanuaryJanuary|November|Nov|11)",r"November", x) )
main_df.loc[:,"birthmonth"] = main_df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"(JanuaryFebruary|December|Dec|12)",r"December", x) )
```

```python
#birthyear

main_df.birthyear = main_df.birthyear.astype(str)
main_df.loc[:,"birthyear"] = main_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^97",r"1997", x) )
main_df.loc[:,"birthyear"] = main_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^98",r"1998", x) )
main_df.loc[:,"birthyear"] = main_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^99",r"1999", x) )
main_df.loc[:,"birthyear"] = main_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^00",r"2000", x) )
main_df.loc[:,"birthyear"] = main_df.loc[:,"birthyear"].apply(lambda x: re.sub(r"^01",r"2001", x) )
```

```python
#course

main_df.loc[:,"course"] = main_df.loc[:,"course"].apply(lambda x: re.sub(r"(^Legal Management$|^BS LM$|^LM$)",r"BS Legal Management", x) )
main_df.loc[:,"course"] = main_df.loc[:,"course"].apply(lambda x: re.sub(r"(^Management Engineering$|^BS ME$|^ME$)",r"BS Management Engineering", x) )
main_df.loc[:,"course"] = main_df.loc[:,"course"].apply(lambda x: re.sub(r"(^Information Technology Entrepreneurship$|^BS ITE$|^ITE$)",r"BS IT Entrepreneurship", x) )
main_df.loc[:,"course"] = main_df.loc[:,"course"].apply(lambda x: re.sub(r"(^Management$|^BS MGT$|^MGT$)",r"BS Management", x) )
main_df.loc[:,"course"] = main_df.loc[:,"course"].apply(lambda x: re.sub(r"(^Communications Technology$|^BS CTM$|^CTM$)",r"BS Communications Technology", x) )
```

```python
#phonenumbers

main_df.loc[:,"phonenumber"] = main_df.loc[:,"phonenumber"].apply(lambda x: re.sub(r"^9",r"+63 9", x) )
```

```python
#carrier

carrier = {"+63 917":"Globe","+63 918":"Smart", "+63 927":"Globe"}
main_df["carrier"] = main_df.loc[:,"phonenumber"].str[0:7].apply(lambda x: carrier[x])
```

```python
#converting to csv
main_df.to_csv('cleanrecweekdata3b.csv', sep=',', index=False)
```

# Required Summary Statistics

#### These were the questions from 3A:

1. How many people were born in each month?
2. How many people were born on each day?
3. How many people were born in each year?
4. How many people come from each course?
5. How many people have numbers from Globe (0917 or 0927) and Smart (0918)? *Note: it's up to you how to format this output*

#### Additional Questions:
* How many Globe subscribers are: Male? Female?
* How many Smart subscribers are: Male? Female?

Answer these questions in an html file called **stats3b.html**. The format of the html is as follows (in template form, so you can use this as format string in your program, though feel free to modify with variable names if you find it easier to do so):


```python
#How many people were born in each month?
birthmonth_count = main_df.groupby("birthmonth").count()

#How many people were born on each day?
birthday_count = main_df.groupby("birthday").count()

#How many people were born in each year?
birthyear_count = main_df.groupby("birthyear").count()

#How many people come from each course?
course_count = main_df.groupby("course").count()

#How many people have numbers from Globe (0917 or 0927) and Smart (0918)?
carrier_count = main_df.groupby("carrier").count()

#How many Globe subscribers are: Male? Female?
#How many Smart subscribers are: Male? Female?
gender_carrier_count = main_df.groupby(["gender","carrier"]).count()

#to test: birthmonth_count["firstname"]
```

```python
#make the data set into a dictionary
birthmonth_dict = birthmonth_count.to_dict()
birthday_dict = birthday_count.to_dict()
birthyear_dict = birthyear_count.to_dict()
course_dict = course_count.to_dict()
carrier_dict = carrier_count.to_dict()
gender_carrier_dict = gender_carrier_count.to_dict()

#only consider the data in the first column (all the other columns have the same data anyway)
birthmonth_counter = birthmonth_dict['firstname']
birthday_counter = birthday_dict['firstname']
birthyear_counter = birthyear_dict['firstname']
course_counter = course_dict['firstname']
carrier_counter = carrier_dict['firstname']
gender_carrier_counter = gender_carrier_dict['firstname']
```

```python
#birthmonth

#sorting the months in order
birthmonths = ['January','February','March','April','May','June','July','August','September','October','November','December']
sorted_birthmonth_counter = dict()
for month in birthmonths:
    sorted_birthmonth_counter[month]=birthmonth_counter[month]

birthmonth_html = """"""
for month in sorted_birthmonth_counter:
    birthmonth_html += "<div class = 'div-data'>" + month + "</div>"
    
birthmonthdata_html = """"""
for month in sorted_birthmonth_counter:
    birthmonthdata_html += "<div class = 'div-data'>" + str((sorted_birthmonth_counter[month])) + "</div>"
```

```python
#birthyear

birthyear_html = """"""
for year in birthyear_counter:
    birthyear_html += "<div class = 'div-data'>" + str(year) + "</div>"
    
birthyeardata_html = """"""
for year in birthyear_counter:
    birthyeardata_html += "<div class = 'div-data'>" + str((birthyear_counter[year])) + "</div>"
```

```python
#birthday

birthday_html = """"""
for day in birthday_counter:
    birthday_html += "<div class = 'div-data'>" + str(day) + "</div>"
    
birthdaydata_html = """"""
for day in birthday_counter:
    birthdaydata_html += "<div class = 'div-data'>" + str((birthday_counter[day])) + "</div>"
```

```python
#course

course_html = """"""
for course in course_counter:
    course_html += "<div class = 'div-data'>" + course + "</div>"
    
coursedata_html = """"""
for course in course_counter:
    coursedata_html += "<div class = 'div-data'>" + str((course_counter[course])) + "</div>"
```

```python
#carrier

carrier_html = """"""
for carrier in carrier_counter:
    carrier_html += "<div class = 'div-data'>" + carrier + "</div>"
    
carrierdata_html = """"""
for carrier in carrier_counter:
    carrierdata_html += "<div class = 'div-data'>" + str((carrier_counter[carrier])) + "</div>" 
```

```python
#carrier by gender

#making a dictionary for the subscribers by gender so that it can be converted to "reference-able" key value pairs according to gender and carrier
subscribers_by_gender = dict()

male_globe = gender_carrier_counter[('Male','Globe')]
female_globe = gender_carrier_counter[('Female','Globe')]
male_smart = gender_carrier_counter[('Male','Smart')]
female_smart = gender_carrier_counter[('Female','Smart')]

subscribers_by_gender['Male Globe Subscribers'] = male_globe
subscribers_by_gender['Female Globe Subscribers'] = female_globe
subscribers_by_gender['Male Smart Subscribers'] = male_smart
subscribers_by_gender['Female Smart Subscribers'] = female_smart

subscribersbygender_html = """"""
for subscriber in subscribers_by_gender:
     subscribersbygender_html += "<div class = 'div-data'>" + subscriber + "</div>"
        
subscribersbygenderdata_html = """"""
for subscriber in subscribers_by_gender:
     subscribersbygenderdata_html += "<div class = 'div-data'>" + str(subscribers_by_gender[subscriber]) + "</div>"      
```
